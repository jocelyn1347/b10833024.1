```python
import matplotlib.pyplot as plt
import pandas as pd
tdata = pd.read_csv("cases_add_Taiwan.csv", encoding='big5')
tdata
```

```python
tdata[['Date','Worldwide']].plot(kind = 'line',x='Date',y='Worldwide',title="COVID-19 Daily Report",legend=False)
plt.show()
```

```python
spots= tdata
spots_pivot=spots.pivot_table(index='Date')
spots_pivot.plot(title='COVID-19 Situation Report',legend=False)
plt.show()
```

```python
kspot=tdata[['Date','Worldwide','Taiwan','Australia']]
kspot.plot(title="COVID-19 Situation Report",legend=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
cntry1_m=tdata['Date']
cntry1_n=tdata['Worldwide']
plt.plot(cntry1_m,cntry1_n,label="WORLDWIDE")
cntry2_m=tdata['Date']
cntry2_n=tdata['Taiwan']
plt.plot(cntry2_m,cntry2_n,lw=2,ls="--",label="Taiwan")
cntry3_m=tdata['Date']
cntry3_n=tdata['Australia']
plt.plot(cntry3_m,cntry3_n,color="Green",lw=2,ls="-.",label="Australia")
plt.legend()
plt.title("COVID-19 Situation Report")
```
